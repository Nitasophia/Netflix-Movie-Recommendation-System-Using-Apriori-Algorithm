# Data Preparation

## 1) Memanggil Library pandas
Pandas kita gunakan untuk membaca dataset yang bentuknya dalam format â€œcsv

```python
import pandas as pd
```

## 2) Memanggil Dataset
Terdapat 4 data yaitu combined_data_1, combined_data_2, combined_data_3, combined_data_4 yang berisi userId, rating, year dan terdapat 1 data yaitu movie_titles yang berisi movie_id, year_of_release, title.

```python
data1 = pd.read_csv('combined_data_1.csv')
data1.columns = ["userId", "rating", "year"]
data1.head()
```

```python
data2 = pd.read_csv('combined_data_2.csv')
data2.columns = ["userId", "rating", "year"]
data2.head()
```

```python
data3 = pd.read_csv('combined_data_3.csv')
data3.columns = ["userId", "rating", "year"]
data3.head()
```

```python
data4 = pd.read_csv('combined_data_4.csv')
data4.columns = ["userId", "rating", "year"]
data4.head()
```

```python
data5 = pd.read_csv('movie_titles.csv', encoding='ISO-8859-1')
data5.head(10)
```

Untuk menghapus kolom yang berisi NaN pada data 5 maka :

```python
data5 = pd.read_csv('movie_titles.csv', encoding='ISO-8859-1')
columns = ['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5']
data5.drop(columns, inplace=True, axis=1)
data5.columns = ['movie_id', 'year_of_release', 'title']
data5.head(10)
```

## 3) Merge Dataset
datamerge1 merupakan gabungan dari combined_data_1 dan combined_data_2

```python
datamerge1 = pd.merge(data1, data2)
datamerge1.head()
```

```python
datamerge1.tail()
```

datamerge2 merupakan gabungan dari combined_data_3 dan combined_data_4

```python
datamerge2 = pd.merge(data3, data4)
datamerge2.head()
```

datamerge3 merupakan gabungan dari combined_data_1 hingga combined_data_4

```python
datamerge3 = pd.merge(datamerge1, datamerge2)
datamerge3.head()
```

```python
datamerge3.tail()
```

Untuk menyimpan dataset yang  sudah diubah sesuai kebutuhan menjadi file CSV yang baru (export datamerge3)

```python
datamerge3.to_csv("datamerge3.csv", index=False)
```

Alldata merupakan gabungan dari kelima data yang dibutuhkan yaitu  combined_data_1, combined_data_2, combined_data_3, combined_data_4 dan movie_titles

```python
alldata = pd.concat([datamerge3,data5],axis=1)
alldata.head()
```

Untuk menyimpan dataset yang sudah diubah sesuai kebutuhan menjadi file CSV yang baru (export alldata)

```python
alldata.to_csv("alldata.csv", index=False)
```
